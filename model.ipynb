```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
import pandas as pd
from nltk.tokenize import sent_tokenize
from nltk import pos_tag, ne_chunk
from datetime import datetime

def extract_events(article_text):
    if not isinstance(article_text, str):
        article_text = str(article_text)
    events = []
    sentences = sent_tokenize(article_text)
    for sentence in sentences:
        words = pos_tag(sentence.split())
        named_entities = ne_chunk(words)
        for entity in named_entities:
            if isinstance(entity, tuple):
                if entity[1] in ['NN', 'NNP'] and entity[0].lower() not in stopwords.words('english'):
                    events.append(entity[0])
    return events

def create_timeline(df):
    timeline = {}
    for index, row in df.iterrows():
        article_text = row['articleBody']
        events = extract_events(article_text)
        if events:
            date = row['Date'] or row['scrapedDate']
            if date:
                if isinstance(date, str):
                    if len(date) == 24 and date[-1] == 'Z':
                        date = datetime.strptime(date[:-1], '%Y-%m-%dT%H:%M:%S')
                    else:
                        try:
                            date = datetime.strptime(date, '%Y-%m-%d')
                        except ValueError:
                            continue
                if isinstance(date, int):
                    date = str(date)
                if date not in timeline:
                    timeline[date] = []
                timeline[date].extend(events)
    return timeline

timeline = create_timeline(df)

timeline = create_timeline(df)

sorted_timeline = sorted(timeline.items(), key=lambda x: x[0])

for date, events in sorted_timeline:
    if isinstance(date, str):
        if len(date) <= 2:
            current_date = datetime.now()
            year_month = current_date.strftime('%Y-%m')
            date_str = f"{year_month}-{date}"
            date = datetime.strptime(date_str, '%Y-%m-%d')
        else:
            date = datetime.strptime(date, '%Y-%m-%d')
    print(date.strftime('%Y-%m-%d'), ":")
    for event in events:
        print(" -", event)
```
