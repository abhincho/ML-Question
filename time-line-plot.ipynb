# Ranking of articels
1. after Grouping By date i have used gpt-neo-125M to rank the articels present that day
2. i have tried different promts to generate
3. i have tried few - shot learning on smaller scale improving few - shot model will give better results due to time constrain i have gone with a direct promt

```python
from transformers import GPT2Tokenizer, GPTNeoForCausalLM
import torch
import torch.nn as nn

# Read the DataFrame
df_israel_war = pd.read_csv('/kaggle/input/warset/war_articels.csv')

# Keep only the first 10 rows
df_israel_war =df_israel_war.dropna(subset=['date'])
df_israel_war['date'] = df_israel_war['date'].str[:10]

# Load tokenizer and model onto GPU
device = torch.device('cuda')

# Set up the tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained('EleutherAI/gpt-neo-125M')
model = GPTNeoForCausalLM.from_pretrained('EleutherAI/gpt-neo-125M')
model.to(device)

# Add padding token
tokenizer.pad_token = tokenizer.eos_token

# Wrap the model with DataParallel
model = nn.DataParallel(model)

# Group articles by date
grouped_articles = df_israel_war.groupby('date')

# Iterate over date groups
for date, group_df in grouped_articles:
    # Function to rank articles for this date group
    def rank_articles(date_group):
        articles = date_group['title']
        prompts = ["Rank the following articles based on their relevance to Israel or war:\n\n" + '\n\n'.join(articles)]
        
        inputs = tokenizer(prompts, return_tensors="pt", truncation=True, padding=True).to(device)
        
        # Access the underlying model's generate method
        with torch.no_grad():
            outputs = model.module.generate(**inputs, max_length=1024, num_return_sequences=1)
        
        ranked_texts = tokenizer.decode(outputs[0], skip_special_tokens=True)
        ranked_list = ranked_texts.split('\n')
        
        ranking = {title: rank for rank, title in enumerate(ranked_list)}
        date_group['rank'] = date_group['title'].map(ranking)
        
        return date_group.sort_values(by='rank')
    
    # Apply ranking function to the current date group
    ranked_df = rank_articles(group_df)
    
    # Save the ranked articles for this date group to CSV
    ranked_df.to_csv(f'ranked_articles_{date}.csv', index=False)
```

```python
import os
import pandas as pd

# Directory containing CSV files
directory = '/kaggle/working/'

# Initialize an empty list to store DataFrames
dfs = []

# Iterate over CSV files in the directory 
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        filepath = os.path.join(directory, filename)
        
        # Read the first row of the CSV file
        df = pd.read_csv(filepath, nrows=1)
        
        # Append the DataFrame to the list
        dfs.append(df)

# Concatenate all DataFrames into a final DataFrame
final_df = pd.concat(dfs, ignore_index=True)

# Save the final DataFrame to a new CSV file
final_df.to_csv('/kaggle/working/combined_data.csv', index=False)
```
