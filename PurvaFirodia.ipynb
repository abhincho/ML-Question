```python
import json
from datetime import datetime
```

```python
def extract_events(events):
    timeline = []
    for event in events:
        # Extract and format the date
        date = event.get("dateModified", {}).get("$date", "")
        if date:
            date = datetime.strptime(date, "%Y-%m-%dT%H:%M:%S.%fZ").strftime("%d %B %Y")
        
        # Extract the title and a brief detail from the article body
        title = event.get("title", "No Title")
        article_body = event.get("articleBody", "No Details")
        first_sentence = article_body.split('.')[0] + '.'
        
        # Create the event summary
        event_summary = {
            "date": date,
            "event": title,
            "details": first_sentence
        }
        
        timeline.append(event_summary)
    
    return timeline
```

```python
def format_timeline_summary(timeline):
    summary_list = []
    
    for event in timeline:
        date = event["date"]
        event_title = event["event"]
        event_details = event["details"]
        
        # Create a formatted summary for the event
        event_summary = f"{date}\n- Event: {event_title}\n- Details: {event_details}"
        
        # Add the event summary to the list
        summary_list.append(event_summary)
    
    return summary_list
```

```python
def read_json_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)
    return data
```

```python
file_path = 'news_article.json'

events = read_json_file(file_path)

timeline = extract_events(events)

timeline_summary = format_timeline_summary(timeline)
```

```python
for event in timeline_summary:
    print(event)
    print("\n")

# Save the summary to a file
with open("timeline_summary.txt", "w", encoding='utf-8') as file:
    for event in timeline_summary:
        file.write(event + "\n\n")
```
